```python
#libraries

import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8) #adjusting the spot of graphs
```

```python
#reading the data

df = pd.read_csv(r'C:\Users\Aras\Desktop\Python\Correlation\movies.csv')
df
```

```python
#Number of null columns
for c in df.columns:
    missing_c = df[c].isnull().sum()
    print('{} - {}'.format(c, missing_c))
    
print('\n#############\n')

#Percentage of null columns
for col in df.columns:
    pct_missing = np.mean(df[col].isna())
    print('{} - {}%'.format(col, round(pct_missing*100)))
```

```python
df.dtypes
```

```python
#Changing float values to integer in order to be efficient

df['budget'] = df['budget'].astype('Int64')
df['gross'] = df['gross'].astype('Int64')
df['votes'] = df['votes'].astype('Int64')
```

```python
#released date and year column are not matching. I want to get the year from released and create another column and delete the old year column
#I want to split the released date and country as released country

df['released'] 
df['released'].str.rsplit(',',2)
df['released'].str[1:4]
```

```python
#Sorting the data to analyse

df.sort_values(by=['score'], inplace=False, ascending=False)
```

```python
pd.set_option('display.max_rows',None)
pd.set_option('display.max_columns',None)
```

```python
#Dropping duplicate values

df.drop_duplicates()
```

```python
#scatter plot with budget and gross

plt.scatter(x=df['budget'],y=df['gross'])
plt.title('Budget versus Gross Earnings')
plt.xlabel('Gross')
plt.ylabel('Budget')
plt.show()
```

```python
#Regression plot budget vs gross

sns.regplot(x='budget',y='gross',data=df, scatter_kws={"color":"orange"},line_kws={"color":"blue"})
```

```python
#Correlation matrixes

df.corr(method='pearson') #default correlation
```

```python
corr_matrix = df.corr() #pearson

sns.heatmap(corr_matrix, annot = True)
plt.title('Correlation Matrix')
plt.show()
```

```python
#changing the data type to give numeric values to every string column

df_numerized = df

for col in df_numerized.columns:
    if df_numerized[col].dtype=='object' :
        df_numerized[col] = df_numerized[col].astype('category')
        df_numerized[col] = df_numerized[col].cat.codes
        
df_numerized
```

```python
df_numerized.corr()
```

```python
#correlation values in linear way
corr_mat = df_numerized.corr()
corr_mat_unstack = corr_mat.unstack()
corr_mat_unstack
```

```python
#Matched correlations
corr_mat_sorted = corr_mat_unstack.sort_values()
corr_mat_sorted
```

```python
# Correlations with high values
corr_mat_sorted[(corr_mat_sorted) > 0.5]
```
